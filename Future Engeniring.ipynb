## Outlier Engineering

```python
import seaborn as sns

# for Q-Q plots
import scipy.stats as stats
```

```python
def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)
    plt.title('Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
diagnostic_plots(ds, 'L7Protocol')
```

```python
diagnostic_plots(ds, 'Destination.Port')
```

```python
def find_skewed_boundaries(df, variable, distance):

    # Let's calculate the boundaries outside which sit the outliers
    # for skewed distributions

    # distance passed as an argument, gives us the option to
    # estimate 1.5 times or 3 times the IQR to calculate
    # the boundaries.

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary
```

#### Ця функція призначена для знаходження меж аномалій для змінної


```python
lead_time_upper_limit, lead_time_lower_limit = find_skewed_boundaries(ds, 'L7Protocol', 1.5)
lead_time_upper_limit, lead_time_lower_limit
```

```python
avg_price_per_room_upper_limit, avg_price_per_room_lower_limit = find_skewed_boundaries(ds, 'Destination.Port', 1.5)
avg_price_per_room_upper_limit, avg_price_per_room_lower_limit
```
